```python
###  Remove trailing spaces from text fields #####
df = df.apply(lambda x: x.str.rstrip() if x.dtype == "object" else x)
```

```python
df.to_csv(r"E:\User\Desktop\sham\DATA SCIENCE\Project1 - Retail_Order_Data_Analysis\orders.csv",index=False)
```

```python
#### Derive new columns discount,sale_price,profit ####
df['discount'] = df['list_price'] * df['discount_percent'] / 100
```

```python
df['sale_price'] = df['list_price'] - df['discount']
```

```python
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
#######  Splitting the dataframes into two dataframes  ########
orders_df = df[['order_id','order_date','ship_mode','segment','country','city','state','postal_code','region']]
sales_df = df[['order_id','category','sub_category','product_id','cost_price','list_price','quantity','discount_percent','discount','sale_price','profit']]
```
